```python
import pandas as pd

df = pd.read_csv("/Users/paigemacmillan/Documents/Fall 2025/MATH 250/Project/vdh_overdose_employment_population_full.csv")
selected_cols = [
    "Overdose Death Year",
    "Overdose Death Geography Level",
    "health_district",
    "Overdose Death Rate per 100,000 Residents",
    "year",
    "unemployment_rate",
    "population"
]

cols_present = [c for c in selected_cols if c in df.columns]
filtered_df = df[cols_present]

stats = filtered_df.describe(include = 'all')
if 'unique' in stats.index:
    stats = stats.drop(index = 'unique')
if 'top' in stats.index:
    stats = stats.drop(index = 'top')
if 'freq' in stats.index:
    stats = stats.drop(index = 'freq')
if 'year' in stats.index:
    stats = stats.drop(index = 'year')

stats.to_csv("/Users/paigemacmillan/Documents/Fall 2025/MATH 250/Project/overdose_descriptivestats.csv")
print("Data Table with Overdose Descriptive Statistics:\n", stats.to_string())
```

```python
## Overdose Death Rate by 100,000 Residents

filtered_df = df[[
    "health_district",
    "Overdose Death Rate per 100,000 Residents"
]]

filtered_df = filtered_df[filtered_df["health_district"] != "Unknown"]
deathrate_stats = filtered_df.groupby('health_district')["Overdose Death Rate per 100,000 Residents"].describe()
deathrate_stats_sorted = deathrate_stats.sort_values(by='mean', ascending=False)
deathrate_stats_sorted = deathrate_stats_sorted.round(3)

deathrate_stats_sorted.to_csv("/Users/paigemacmillan/Documents/Fall 2025/MATH 250/Project/deathrate_descriptivestats.csv")
print("Overdose Death Rate per 100,000 Residents Descriptive Statistics:\n", deathrate_stats_sorted.to_string())
```

```python
## By Unemployment Rate

filtered_df = df[[
    "health_district",
    "unemployment_rate"
]]

filtered_df = filtered_df[filtered_df["health_district"] != "Unknown"]
unemployment_stats = filtered_df.groupby('health_district')["unemployment_rate"].describe()
unemployment_stats_sorted = unemployment_stats.sort_values(by='mean', ascending=False)
unemployment_stats_sorted = unemployment_stats_sorted.round(3)

unemployment_stats_sorted.to_csv("/Users/paigemacmillan/Documents/Fall 2025/MATH 250/Project/unemployment_descriptivestats.csv")
print("Unemployment Rate Descriptive Statistics:\n", unemployment_stats_sorted.to_string())
```

```python
## By population

filtered_df = df[[
    "health_district",
    "population"
]]

filtered_df = filtered_df[filtered_df["health_district"] != "Unknown"]
population_stats = filtered_df.groupby('health_district')["population"].describe()
population_stats_sorted = population_stats.sort_values(by='mean', ascending=False)
population_stats_sorted = population_stats_sorted.round(3)

population_stats_sorted.to_csv("/Users/paigemacmillan/Documents/Fall 2025/MATH 250/Project/population_descriptivestats.csv")
print("Population Descriptive Statistics:\n", population_stats_sorted.to_string())
```
